```python
g = text(1, (0,0), color='red'); d = dict()
```

```python
for i in range(32):
    d[(i,0)]=1
    d[(0,i)]=1
```

```python
for i in range(32-1):
    for j in range(i+1):
        try:
            d[(i-j+1, j)] = d[(i-j+1-1, j)] + d[(i-j+1, j-1)]                
        except KeyError:
            d[(i-j+1-1, j)] = 1
            d[(i-j+1, j-1)] = 1
        if d[(i-j+1, j)] % 2 == 0:
            g += text(d[(i-j+1, j)],(i-j+1, j))
        else:
            g += text(d[(i-j+1, j)],(i-j+1, j), color='red')
    for j in range(i+1):
        if d[(i-j, j+1)] % 2 == 0:
            g += text(d[(i-j, j+1)],(i-j, j+1))
        else:
            g += text(d[(i-j, j+1)],(i-j, j+1), color='red')
```

```python
g.show(axes=False)
```
